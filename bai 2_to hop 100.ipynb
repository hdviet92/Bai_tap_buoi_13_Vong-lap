```python
n = 100

maxx = n//2
maxy = n//3
maxz = n//5
count = 0

for x in range(maxx+1):
    for y in range(maxy+1):
        for z in range(maxz+1):
            if 2*x + 3*y + 5*z == 100:
                print (x, y ,z)
                count += 1
print('Có {count} cách đổi'.format(count=count))
    
    
    
```
